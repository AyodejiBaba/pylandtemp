```python
import numpy as np
from PIL import Image
```

```python
from emissivity.emissivity import Emissivity
```

```python
import data
from general_utils import compute_ndvi
import matplotlib.pyplot as plt
```

## Image paths

```python
path_red = './data/espirito_santo/LC08_L1TP_216074_20160531_20180528_01_T1_B4.TIF'
path_ir = './data/espirito_santo/LC08_L1TP_216074_20160531_20180528_01_T1_B5.TIF'
```

## Load read and infrared images

```python
red_band = Image.open(path_red)
ir_band = Image.open(path_ir)
```

```python
red_band_array = np.array(red_band)
ir_band_array = np.array(ir_band)
```

```python
plt.imshow(red_band_array)
plt.colorbar()
```

## Compute NDVI

```python
ndvi = compute_ndvi(ir_band_array, red_band_array)
```

```python
plt.imshow(ndvi)
plt.title('NDVI')
plt.colorbar()
```

## Compute Emissivity

#### 'avdan' method

```python
emm_test_avdan = Emissivity(ndvi, None)('avdan')
plt.imshow(emm_test_avdan)
plt.colorbar()
```

```python
emm_test_gopinadh = Emissivity(ndvi, red_band_array)('gopinadh')
plt.imshow(emm_test_gopinadh)
plt.colorbar()
```

```python
emm_test_xiaolei = Emissivity(ndvi, red_band_array)('xiaolei')
plt.imshow(emm_test_xiaolei)
plt.colorbar()
```
